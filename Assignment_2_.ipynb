```python

import pandas as pd
```

```python

df = pd.read_csv('/content/employee_performance.csv')
print("Dataset Head:")
print(df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Display basic statistics
print("Summary Statistics:")
print(df.describe())
```

```python

plt.figure(figsize=(8, 5))
sns.histplot(df['Monthly_Salary'], bins=10, kde=True, color='green')
plt.title("Distribution of Monthly salaries")
plt.xlabel("Employee_ID")
plt.ylabel("Monthly_Salary")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['Years_of_Experience'], y=df['Performance_Score'], hue=df['Department'])
plt.title("Years_of_Experience vs. Performance_Score")
plt.xlabel("Years_of_Experience")
plt.ylabel("Performance_Score")
plt.show()
```
